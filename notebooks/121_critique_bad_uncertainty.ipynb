# Critique Experiment 2 (Strict Pipeline): Higher Bad-Regime Uncertainty

Methodology aligned with notebook 116 (script-first, notebook-display).

```python
import os
import sys
import shlex
import pathlib
import subprocess

import pandas as pd
import torch
from IPython.display import Image, Markdown, display


def _find_project_root() -> pathlib.Path:
    here = pathlib.Path.cwd().resolve()
    for p in [here, *here.parents]:
        if (p / "src").is_dir() and (p / "scripts" / "build_critique_figures_author_strict.py").is_file():
            return p
    cand = pathlib.Path("/content/econml")
    if (cand / "src").is_dir() and (cand / "scripts" / "build_critique_figures_author_strict.py").is_file():
        return cand
    raise RuntimeError("Project root not found.")


PROJECT_ROOT = _find_project_root()
os.chdir(PROJECT_ROOT)
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

ARTIFACTS_ROOT = str(PROJECT_ROOT / "artifacts")
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
DTYPE = "float64"
USE_SELECTED = True
CONS_MODE = "author"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DEVICE:", DEVICE)
```

```python
FIG_DIR = str(PROJECT_ROOT / "artifacts" / "critique_figures_nb121_author_strict")
os.makedirs(FIG_DIR, exist_ok=True)

cmd = [
    sys.executable,
    "scripts/build_critique_figures_author_strict.py",
    "--artifacts-root", ARTIFACTS_ROOT,
    "--fig-dir", FIG_DIR,
    "--device", DEVICE,
    "--dtype", DTYPE,
    "--use-selected", str(USE_SELECTED).lower(),
    "--cons-mode", CONS_MODE,
    "--ensure-postprocess", "true",
    "--ensure-ir", "true",
    "--scenarios", "bad_uncertainty",
]
print("Running:", " ".join(shlex.quote(x) for x in cmd))
subprocess.run(cmd, cwd=str(PROJECT_ROOT), check=True)
print("Done. FIG_DIR=", FIG_DIR)
```

```python
for pol in ["taylor", "discretion", "commitment"]:
    p = pathlib.Path(FIG_DIR) / f"critique_bad_uncertainty_{pol}.png"
    if p.exists():
        display(Markdown(f"### Bad Uncertainty: {pol}"))
        display(Image(filename=str(p), width=980))
    else:
        print("Missing figure:", p)

p_m = pathlib.Path(FIG_DIR) / "critique_bad_uncertainty_moments.csv"
p_p = pathlib.Path(FIG_DIR) / "critique_bad_uncertainty_pivot.csv"
if p_m.exists():
    display(Markdown("### Moments"))
    display(pd.read_csv(p_m))
else:
    print("Missing:", p_m)
if p_p.exists():
    display(Markdown("### Pivot"))
    display(pd.read_csv(p_p))
else:
    print("Missing:", p_p)
```
